```python
import numpy as np
from sympy import *
import matplotlib.pyplot as plt
from scipy.integrate import quad

p=147/100
w=np.pi

def periodically_continued(a, b):
    interval = b - a
    return lambda f: lambda x: f((x - a) % interval + a)

@periodically_continued(-1, 1)
def s(t):
  return p*t**2;

def s1(t):
  return s(t)*np.cos(w*t)

def s2(t):
  return s(t)*np.cos(2*w*t)

# call quad to integrate f from -2 to 2
a0 = 1/2*quad(s, -1, 1)[0]
a1 = quad(s1, -1, 1)[0]
a2 = quad(s2, -1, 1)[0]

print("a0=", a0)
print("a1=", a1)
print("a2=", a2)

def fourier(t):
  return a0+a1*np.cos(np.pi*t)+a2*np.cos(2*np.pi*t)

t = np.linspace(-3, 3, 100)
fig, ax = plt.subplots()
plt.xlabel("czerwony - s(t)\nniebieski - fourier")
plt.title("Wykres funkcji s dla wielu okres√≥w i aproksymacji Fouriera")
ax.plot(t, fourier(t), color='blue')
ax.plot(t, s(t), color='red')
ax.grid(True)
plt.show()

```
